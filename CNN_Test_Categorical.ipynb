```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

path='/content/drive/MyDrive/Colab Notebooks/Proyecto/prueba/prueba8.jfif'
img=image.load_img(path, target_size=(150, 150))
x=image.img_to_array(img)
image=np.expand_dims(x, axis=0)
```

```python
import json
a_file = open("/content/drive/MyDrive/Colab Notebooks/Proyecto/Modelo entrenado 2/fungi.json", "r")
#a_file = open("/content/drive/MyDrive/Colab Notebooks/Proyecto/fungi_indices.json", "r")
class_indices = json.load(a_file)
model = load_model('/content/drive/MyDrive/Colab Notebooks/Proyecto/Modelo entrenado 2/fungi_categorical.h5')
#model = load_model('/content/drive/MyDrive/Colab Notebooks/Proyecto/fungi_categorical.h5')
classes = model.predict(image)
print(classes)
clas =np.where(classes[0] == max(classes[0]))
categoria = clas[0][0]
print(categoria)
print(list(class_indices.keys())[list(class_indices.values()).index(categoria)])
type1= list(class_indices.keys())[list(class_indices.values()).index(categoria)]
```
