```python
# In[9]:
import tensorflow as tf
from tensorflow.keras import Model

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv3D, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential()
#model.add(Conv3D(32, (3,3), activation='relu', input_shape=(150, 150, 3)))
#model.add(Conv3D(32, (3,3,1), activation='relu', input_shape=(150, 150, 3)))
#model.add(MaxPooling2D(2, 2))
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(Dropout(0.1))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(12, activation='softmax'))
```

```python
model.summary()
```

```python
# In[11]:
from tensorflow.keras.optimizers import RMSprop, Adam
model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics = ['acc'])
```

```python
# In[12]:
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1.0/255.)
validation_datagen = ImageDataGenerator(rescale = 1.0/255.)
test_datagen = ImageDataGenerator(rescale = 1.0/255.)

train_generator = train_datagen.flow_from_directory(train_dir, batch_size=20, class_mode='categorical', target_size=(150, 150))

validation_generator = validation_datagen.flow_from_directory(validation_dir, batch_size=20, class_mode='categorical', target_size=(150, 150))

test_generator = test_datagen.flow_from_directory(test_dir, batch_size=20, class_mode='categorical', target_size=(150, 150))
```

```python
batch_size = 20
steps_per_epoch = train_generator.n // batch_size
validation_steps = validation_generator.n // batch_size
#validation_generator = 1

print (steps_per_epoch)
print (validation_steps)
```

```python
# In[14]:
history = model.fit(train_generator, 
                    steps_per_epoch = steps_per_epoch, 
                    epochs=25
                    , 
                    validation_data=validation_generator, 
                    validation_steps=validation_steps, 
                    verbose=1)
#history = model.fit(train_generator, steps_per_epoch = steps_per_epoch, epochs=180, validation_data=validation_generator, verbose=2)

#Guardar el modelo entrenado
model.save('/content/drive/MyDrive/Colab Notebooks/Proyecto/fungi_categorical.h5')
```
